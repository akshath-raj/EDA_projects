```python
def print_unique(df,lable):
    len_unique_lable=len(df[lable].unique())
    unique_lable=df[lable].unique()
    print(f'Number of unique {lable} are {len_unique_lable} and they are: {unique_lable}')
```

```python
print_unique(df,'YOB')
print_unique(df,'Area code')
print_unique(df,'Location')
print_unique(df,'Activity')
print_unique(df,'Smoker')
print_unique(df,'Cuisines')
print_unique(df,'Alcohol ')
```
