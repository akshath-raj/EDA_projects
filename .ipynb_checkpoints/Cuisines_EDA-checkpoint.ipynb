```python
def print_unique(df,lable):
    len_unique_lable=len(df[lable].unique())
    unique_lable=df[lable].unique()
    print(f'Number of unique {lable} are {len_unique_lable} and they are: {unique_lable}')
```

```python
print_unique(df,'YOB')
print_unique(df,'Area code')
print_unique(df,'Location')
print_unique(df,'Activity')
print_unique(df,'Smoker')
print_unique(df,'Cuisines')
print_unique(df,'Alcohol ')
```

```python
fig, axs = plt.subplots(5,1, figsize=(15, 25))
axs = axs.flatten()

sns.countplot(data=df,x='Location',ax=axs[0])
sns.countplot(data=df,x='Smoker',ax=axs[1])
sns.countplot(data=df,x='Activity',ax=axs[2])
sns.countplot(data=df,x='Cuisines',ax=axs[3])
sns.countplot(data=df,x='Often A S',ax=axs[4])
```

```python
fig, axs = plt.subplots(5,1, figsize=(15, 20))
axs = axs.flatten()
sns.boxenplot(data=df,x='Location', y="Overall Rating",ax=axs[0])
sns.boxenplot(data=df,x='Gender', y="Overall Rating",ax=axs[1])
sns.boxenplot(data=df,x='Activity', y="Overall Rating",ax=axs[2])
sns.boxenplot(data=df,x='Alcohol ', y="Overall Rating",ax=axs[3])
sns.boxenplot(data=df,x='Cuisines', y="Overall Rating",ax=axs[4])
```

```python
# Inferences
# The data set has reviews from 200 people and does not contain any missing models.
# There are 7 classes of Cuisines: Japanese, Indian, Seafood, Filipino, Italian, Chinese and French.
# The column label is mistakenly labelled as 'Alcohol ' instead of 'Alcohol'
# Although the food rating of Filipino cuisine is lowest, it is the second most commonly ordered cuisine
# People with an average budget tend to go for Indian or Filipino cuisine and people with a high budget go for Japanese or French cuisine.
# Chinese and French cuisine is only ordered by people with a medium or high budget indicating that Chinese and French dishes are costly.
# Those who regularly smoke have given higher service ratings indicating the restaurant might allow smoking inside the premises.
# The older customers have opted more for Italian cuisine.
# Females have given higher overall rating than men
```
